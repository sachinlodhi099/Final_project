```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error
```

```python

df = pd.read_csv("TSLA.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)


data = df[['Adj Close']].values
```

```python

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python

def create_sequences(dataset, time_step=60):
    X, y = [], []
    for i in range(time_step, len(dataset)):
        X.append(dataset[i-time_step:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

time_step = 60
X, y = create_sequences(scaled_data, time_step)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python

rnn_model = Sequential([
    SimpleRNN(100, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    SimpleRNN(100, return_sequences=False),
    Dropout(0.2),
    Dense(50),
    Dense(1)
])
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
```

```python

lstm_model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(50),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
```

```python

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print("\nTraining SimpleRNN model...")
history_rnn = rnn_model.fit(X_train, y_train, validation_data=(X_test, y_test),
                            epochs=50, batch_size=32, callbacks=[early_stop], verbose=1)

print("\nTraining LSTM model...")
history_lstm = lstm_model.fit(X_train, y_train, validation_data=(X_test, y_test),
                              epochs=50, batch_size=32, callbacks=[early_stop], verbose=1)
```

```python

rnn_pred = rnn_model.predict(X_test)
lstm_pred = lstm_model.predict(X_test)
```

```python

rnn_pred_prices = scaler.inverse_transform(rnn_pred)
lstm_pred_prices = scaler.inverse_transform(lstm_pred)
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python

rnn_mse = mean_squared_error(actual_prices, rnn_pred_prices)
lstm_mse = mean_squared_error(actual_prices, lstm_pred_prices)

rnn_val_loss = min(history_rnn.history['val_loss'])
lstm_val_loss = min(history_lstm.history['val_loss'])
```

```python
print("\n Model Performance Comparison:")
print(f"SimpleRNN -> val_loss: {rnn_val_loss:.6f}, MSE: {rnn_mse:.6f}")
print(f"LSTM      -> val_loss: {lstm_val_loss:.6f}, MSE: {lstm_mse:.6f}")
```

```python

if lstm_mse < rnn_mse:
    best_model, best_name, best_pred = lstm_model, "LSTM", lstm_pred_prices
else:
    best_model, best_name, best_pred = rnn_model, "SimpleRNN", rnn_pred_prices

print(f"\n Best model based on MSE: {best_name}")
```

```python
predicted = best_model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python

plt.figure(figsize=(10,6))
plt.plot(actual_prices, color='blue', label='Actual Price')
plt.plot(best_pred, color='red', label=f'Predicted Price ({best_name})')
plt.title(f'TSLA Stock Price Prediction - {best_name}')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python

last_60_days = scaled_data[-time_step:]
future_predictions = []

for _ in range(10):
    pred_input = np.reshape(last_60_days, (1, time_step, 1))
    pred_price = best_model.predict(pred_input)
    future_predictions.append(pred_price[0, 0])
    last_60_days = np.append(last_60_days[1:], pred_price, axis=0)
```

```python
future_prices = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
print("\n Next 10 days predicted prices:\n", future_prices)
```
